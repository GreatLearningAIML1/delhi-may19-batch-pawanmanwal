```
# 1. Find the mean of old scheme and new scheme column
print('Mean of Old Scheme', titan_df.Old_Scheme.mean())
print('Mean of New Scheme', titan_df.New_Scheme.mean())
```

```
from scipy.stats import ttest_ind,ttest_rel
```

```
# 2. Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs?

# Alternate Hypthosis: The sale figure has improved
# Null Hypthosis: The same figure has not improved
```

```
stat, p = ttest_ind(titan_df.Old_Scheme, titan_df.New_Scheme)
print('Statistics=%.3f, p=%.3f' % (stat, p))
```

```
# 3. What conclusion does the test (p-value) lead to?
# Since p value is greater than .05, so null hypothesis is accepted. That means that new scheme doesn't have incresed sales figure
```

```
# Created a new Data frame having new scheme data
titan_df['New_New_Scheme']=titan_df['New_Scheme'].add(5000)   
```

```
# 4.a) The probability of a type 1 error?
# Probability of Type 1 Error is 5% as per industry standards
```

```
# 4.b) What is the p- value of the hypothesis test if we test for a difference of $5000?
stat2, p2 = ttest_ind(titan_df['New_Scheme'], titan_df['New_New_Scheme'])
print('New Value of Statistics=%.3f, p=%.3f' % (stat2, p2))
```

```
# Again the p value is greater than .05, so accepting the null hyphothesis that is even after increasing the sales break even will not be achieved.
```

```
# 4.c) Power of the test
# Power of test is probability of surely rejecting null hypothesis that is 1-Beta(Type 2 Error)
```

```
from statsmodels.stats.power import ttest_power
```

```
for col in titan_df.columns:
  colN=col + '_Zscore'
  titan_df[colN]=(titan_df[col]-titan_df[col].mean())/titan_df[col].std(ddof=0)
```

```
zStatistics = (titan_df.New_Scheme[0]-titan_df.New_Scheme.mean())/np.std(titan_df['New_Scheme'])
# Power of test for first sample value
ttest_power(zStatistics,nobs=len(np.array(titan_df['New_Scheme']))-1,alpha=0.05,alternative='larger')
```
