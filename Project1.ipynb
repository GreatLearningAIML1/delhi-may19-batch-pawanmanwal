```
import pandas as pd
import numpy as np
```

```
laliga=pd.read_csv('/content/drive/My Drive/Colab Notebooks/Project1/Laliga.csv')
laliga.head(10)
```

```
# All the dash has been replaced with 0 to carry out arithematic operations
laliga=laliga.replace('-', 0);
laliga.head(2)
```

```
# Since debut column has year range as well, so the starting year has been picked and can be used now for further operations
laliga.Debut=laliga['Debut'].str.split('-').str[0]
```

```
# Converting the Debut column to numeric to perform arithmetic operation
laliga['Debut'] = pd.to_numeric(laliga['Debut'])

# Printing all the rows for which year range is 1930 to 1980 , both the years including
laliga[(laliga['Debut']>=1930) & (laliga['Debut']<=1980)]
```

```
# Converting the data type of Points column from Object to Integer to perform nlargest as nlargest doesn't work on object dtypes
laliga['Points'] = pd.to_numeric(laliga['Points'])
# Display all the rows
print(laliga.nlargest(5, ['Points']))

#Display the list of teams
print(laliga.nlargest(5, ['Points']).Team)
```

```
def Goal_diff_count(laliga):
  laliga['Diff']=laliga['GoalsFor']-laliga['GoalsAgainst']
  return laliga
```

```
# Converting the data type of GoalsFor and GoalsAgainst column from Object to Integer to perform arithmetic operations
laliga['GoalsFor'] = pd.to_numeric(laliga['GoalsFor'])
laliga['GoalsAgainst'] = pd.to_numeric(laliga['GoalsAgainst'])
laliga = Goal_diff_count(laliga)
```

```
# I couldn't understand how we use the same function to get the max and min differences.
# However, I have update the diff column in the main df

# Team having maximum difference in goals
laliga[laliga['Diff']==laliga.Diff.max()]
```

```
# Team having minimum difference in goals
laliga[laliga['Diff']==laliga.Diff.min()]
```

```
laliga['GamesWon'] = pd.to_numeric(laliga['GamesWon'])
laliga['GamesPlayed'] = pd.to_numeric(laliga['GamesPlayed'])
```

```

# Seperate column created for Winning Percentage in the same data frame
laliga['Winning Percent']=(laliga.GamesWon/laliga.GamesPlayed)*100
```

```
# replacing numeric errors with 0
laliga['Winning Percent']=laliga['Winning Percent'].replace(np.nan,0)
laliga
```

```
# Displaying Top 5 teams in order of their winning percentage
laliga.nlargest(5,['Winning Percent'])
```

```
# Grouped them on the basis of Best Position
laliga2=laliga.groupby('BestPosition',as_index=False).sum()

#Display only Best position and points(sum of grouped by)
laliga2.loc[:,['BestPosition','Points']]
```
